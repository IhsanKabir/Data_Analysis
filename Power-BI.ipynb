```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #For visualization
%matplotlib inline
import seaborn as sns
from glob import glob
```

```python
df.dropna(inplace=True)
```

```python
#A bar chart to Interface
alt=sns.countplot(x='Interface', data=df)

for bars in alt.containers:
    alt.bar_label(bars)
    sns.set(rc={'figure.figsize':(17,7)})
```

```python
#plotting a bar for interface and total price

Sales_gen=df.groupby(['Interface'], as_index=False) ['Total Price'].sum().sort_values(by='Interface', ascending=False)
sns.set(rc={'figure.figsize':(27,7)})
sns.barplot(x= 'Interface', y='Total Price' , data =Sales_gen)
```

```python
#Age group based TRNC

ag= sns.countplot(data= df, x='Age Group' , hue='TRNC')

for bars in ag.containers:
    ag.bar_label(bars)
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['A/L'], as_index=False)['Total Price'].sum().sort_values(by='Total Price', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.scatterplot(data = sales_state, x = 'A/L',y= 'Total Price')
```
